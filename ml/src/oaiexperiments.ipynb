```python
from openai import OpenAI
from dotenv import load_dotenv
import os
import constants as c

os.environ.pop("OPENAI_API_KEY", None)
load_dotenv()
apikey = os.getenv("OPENAI_API_KEY")

client = OpenAI(api_key=apikey)
```

```python
def naiveMovieSuggestion(userPrompt, systemPrompt, model="gpt-5-nano"):
    print(f"DEBUG: model is {model}")
    prompt = userPrompt
    response = client.responses.create(
        model=model,
        input=[
            {
                "role": "system",
                "content": systemPrompt,
            },
            {"role": "user", "content": prompt},
        ],
        service_tier="priority",
    )

    return response.output_text
```

```python
naiveMovieSuggestion(
    "a children's movie",
    c.LLM_SYSTEM_PROMPT,
)
```

```python
response = client.responses.create(model="gpt-5-nano", input="a children's movie")

print(response.output_text)
```

```python
def MovieSuggestionWithMinimalReasoningAndMaxToken(
    userPrompt, systemPrompt, model="gpt-5-nano"
):
    print(f"DEBUG: model is {model}")
    prompt = userPrompt
    response = client.responses.create(
        model=model,
        input=[
            {
                "role": "system",
                "content": systemPrompt,
            },
            {"role": "user", "content": prompt},
        ],
        service_tier="priority",
        reasoning={"effort": "minimal"},
        max_output_tokens=1000,
    )

    return response
```

```python
MovieSuggestionWithMinimalReasoningAndMaxToken(
    "a children's movie",
    c.LLM_SYSTEM_PROMPT,
).output_text
```
